<h1>One Hot Encoding</h1>

<h2>Using Pandas</h2>

```python
import pandas as pd
import numpy as np
import sklearn 
```

```python
df = pd.read_csv("cars.csv")
```

```python
df.head(5)
```

```python
df['brand'].value_counts()
```

```python
pd.get_dummies(df, columns = ['owner', 'fuel'])
```

<h3>K-1 Encoding</h3>

```python
pd.get_dummies(df, columns = ['owner', 'fuel'], drop_first = True).sample(5)
```

<h2>Using Sklearn</h2>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1], df.iloc[:, -1], test_size = 0.3, random_state = 0)
X_train.sample(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohc = OneHotEncoder(drop = "first", sparse_output =False)
```

```python
# ohc.fit(X_train.iloc([:, 2:]))
# X_train_fuel_owner = ohc.transform(X_train[])
ohc.fit(X_train[['owner', 'fuel']])
X_train_fuel_owner = ohc.transform(X_train[['owner','fuel']])
X_train_fuel_owner.shape
```

```python
X_train_new = np.hstack((X_train[['brand','km_driven']], X_train_fuel_owner))
X_train_new
```

<h3>OneHotEncoding with only Top Category</h3>

```python
counts = df['brand'].value_counts()
```

```python
threshold = 100
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon'), drop_first = True)
```

<h2>Using Transformers</h2>

```python
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("covid_toy.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
transformer = ColumnTransformer( transformers = [
    ('tf1', SimpleImputer(), ['fever']),
    ('tf2', OrdinalEncoder(categories = [['Mild', 'Strong']]), ['cough']),
    ('tf3', OneHotEncoder(sparse_output = False, drop = "first"), ['gender', 'city'])   
], remainder = 'passthrough')

# transformer = ColumnTransformer(transformers=[
#     ('tnf1',SimpleImputer(),['fever']),
#     ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
#     ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
# ],remainder='passthrough')
```
